# Example Spectral Analysis

```python
from src.waveTools import readDatawell, waveStats
```

## Prepare the data

The toolbox is designed to work with data in 30 minute segments and either Mk4 displacement data (2.56hz) or Mk3 displacement data (1.28hz).

The following data format is assumed when parsing data into the module.

| Time|Heave|North|West|
| ----|-----|-----|----|
| float|float|float|float|

```python
df = readDatawell.readDisplacement('./example_data/brisbane_4183{disp}2020-06-01.csv')
```

Select a subset of the data

```python
df = df['2020-06-01 23:15':]
df.head()
```

```python
len(df)
```

Remove the status field as it does not conform with the data type

```python
df = df.drop(['status'], axis = 1)
df = df.reset_index(drop = True)
df.head()
```

## Using the Module

We just need to call the module and parse through our data object. 

**Note:** if you have Mk4 data (2.56hz) it will be downsampled to 1.28 for the directional spectra calculation

```python
data = waveStats.displacements(df)
```

You can now call a variety of functionality, such as wave statistics and number of plotting functions.

**Tip:** to get a list of functions in Jupyter type you object (e.g. data) and bullet point after it, then PRESS tab

```python
data.wave_stats().T
```

some useful plotting functions have been included within the module

```python
data.plot_spectra()
```

```python
data.plot_directional_spectra()
```
